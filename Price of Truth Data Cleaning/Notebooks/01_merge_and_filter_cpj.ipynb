```python
import pandas as pd
from google.colab import files

# 1) Upload Killed.csv
print("Upload Killed.csv")
uploaded_k = files.upload()
killed_file = list(uploaded_k.keys())[0]

# 2) Upload Imprisoned.csv
print("Upload Imprisoned.csv")
uploaded_i = files.upload()
imprisoned_file = list(uploaded_i.keys())[0]

# 3) Load data
killed = pd.read_csv(killed_file)
imprisoned = pd.read_csv(imprisoned_file)

# 4) Extract year from Date
killed["year"] = pd.to_datetime(killed["Date"], errors="coerce").dt.year
imprisoned["year"] = pd.to_datetime(imprisoned["Date"], errors="coerce").dt.year

# 5) Filter years 2010â€“2024
killed = killed[(killed["year"] >= 2010) & (killed["year"] <= 2024)]
imprisoned = imprisoned[(imprisoned["year"] >= 2010) & (imprisoned["year"] <= 2024)]

# 6) Select and tag columns
killed_clean = killed[[
    "Country",
    "year",
    "Status",
    "Motive",
    "Type of Death",
    "Journalist or Media Worker"
]].copy()
killed_clean["case_status"] = "killed"

imprisoned_clean = imprisoned[[
    "Country",
    "year",
    "Status",
    "Charges",
    "Journalist or Media Worker"
]].copy()
imprisoned_clean["case_status"] = "imprisoned"

# 7) Rename columns
killed_clean = killed_clean.rename(columns={
    "Country": "country",
    "Status": "cpj_status",
    "Motive": "motive",
    "Type of Death": "type_of_death",
    "Journalist or Media Worker": "role"
})

imprisoned_clean = imprisoned_clean.rename(columns={
    "Country": "country",
    "Status": "cpj_status",
    "Charges": "charges",
    "Journalist or Media Worker": "role"
})

# 8) Merge
cpj_master = pd.concat([killed_clean, imprisoned_clean], ignore_index=True)

# 9) Strip text columns
for col in ["country", "cpj_status", "motive", "type_of_death", "charges", "role", "case_status"]:
    if col in cpj_master.columns:
        cpj_master[col] = cpj_master[col].astype(str).str.strip()

# 10) Save and download
output_name = "cpj_master_clean.csv"
cpj_master.to_csv(output_name, index=False)
files.download(output_name)
```
